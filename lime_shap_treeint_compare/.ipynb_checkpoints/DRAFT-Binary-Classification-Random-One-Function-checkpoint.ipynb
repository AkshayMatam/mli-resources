## Ranked Prediction Plot

```python
def ranked_preds_plot(y, valid, preds):
    
    """ Generates ranked prediction plot.
    
    :param y: Name of target variable.
    :param valid: Name of validation H2OFrame.
    :param preds: Column vector of predictions to plot.

    """
    
    # plot top frame values
    preds.columns = ['predict']
    yhat_frame = pd.concat([valid,preds],axis=1)
    print(yhat_frame.loc[0:10, [y, 'predict']])

    # plot sorted predictions
    yhat_frame_df = yhat_frame[[y, 'predict']]
    yhat_frame_df.sort_values(by='predict', inplace=True)
    yhat_frame_df.reset_index(inplace=True, drop=True)
    _ = yhat_frame_df.plot(title='Ranked Predictions Plot')
```

## Signal Generating Function 1

\begin{gather*}
y = x_1*x_4 + |x_8|* (x_9)^2
\end{gather*}

<br>rows:
mod 1 

```python
def print_function1_contribs(row):
    
    num1_contrib = row['num1'].iloc[0]
    print('Expected num1 contribution:', num1_contrib)
    
    num4_contrib = row['num4'].iloc[0]
    print('Expected num4 contribution:', num4_contrib)
    
    num8_contrib = np.abs(row['num8'].iloc[0]) 
    print('Expected num8 contribution:', num8_contrib)
    
    num9_contrib = np.power(row['num9'].iloc[0], 2)
    print('Expected num9 contribution:', num9_contrib)
```

## Standards to decide tree interpreter's result:

<span style="color:green">green</span>:
<br>
for a row to receive a <span style="color:green">green</span> rating, the sign of the tree interpreter result has to match its print_function_contribs, also there will be no other variables than what's expected showing up as biggest positive or negetive contributors 
<br>
<span style="color:red">red</span>:
<br>
For a row to receive a <span style="color:red">red</span> rating, none of the print_function_contribs results show up as the biggest positive or negetive contributors in tree interpreter. 
<br>
<span style="color:yellow">yellow</span>:
<br>
For a row to receive a <span style="color:yellow">yellow</span> rating, some of print_function_contribs results show up in tree interpreter as big contributors. 


**These colors should not be confused with the tree interpreter tables. The colors used here are to judge if the tree interpreter's result matches with our expectation. In the tree interpreter tables, red cells means negetive contributions and green cells are positive contributions. **

# Tree Interpreter on Random with Signal with One Signal Generating Function

\begin{gather*}
y = x_1 * x_4 + |x_8| * (x_9)^2
\end{gather*}

<br>rows:
All

<br>Note:
There is no noise added to the target. Only one transformation function applied to all rows through out the dataset on the target. We expect xgboost to be able to pick up this pattern, and tree interpreter as well. 

Therefore tree interpreter is expected to find num1, num4, num8 and num9 as the biggest positive contributors for all rows. 

```python
ds_ = DataMakerAndGetter(nrows=200000, target='binary', one_function = True)
rsoframe = ds_.make_random_with_signal()
rsotrain, rsovalid, rsotest = rsoframe.split_frame([0.4, 0.3], seed = 12345)
```

```python
y = 'target'
X = [name for name in rsotrain.columns if name not in [y,'row_id','function','cat1','cat2','cat3']]
print(X)
```

```python
rsotrain[y] = rsotrain[y].asfactor()
rsovalid[y] = rsovalid[y].asfactor()
rsotest[y] = rsotest[y].asfactor()
```

```python
rsotrain_pd = rsotrain.as_data_frame()
rsovalid_pd = rsovalid.as_data_frame()
rsotest_pd = rsotest.as_data_frame()
```

```python
rsotrain_dm = xgb.DMatrix(rsotrain_pd[X],
                      rsotrain_pd[y])
rsovalid_dm = xgb.DMatrix(rsovalid_pd[X],
                      rsovalid_pd[y])
rsotest_dm = xgb.DMatrix(rsotest_pd[X])
```

```python
ave_y = rsotrain_pd['target'].mean()
print(ave_y)
```

```python
params = {
     'base_score': ave_y,
     'booster': 'gbtree',
     'colsample_bytree': 0.9,
     'eta': 0.01,
     'eval_metric': 'auc',
     'max_depth': 18,
     'nthread': 4,
     'objective': 'binary:logistic',
     'reg_alpha': 0.01,
     'reg_lambda': 0.001,
     'seed': 12345,
     'silent': 0,
     'subsample': 0.1}


#Best error = 0.996709

watchlist = [(rsotrain_dm, 'train')]

rso_model = xgb.train(params, 
                     rsotrain_dm, 
                     400,
                     evals=watchlist, 
                     verbose_eval=True)
```

```python
rso_preds = pd.DataFrame(rso_model.predict(rsovalid_dm))
rso_preds.head(10)
```

```python
rso_decile_frame = pd.concat([rsovalid_pd, rso_preds], axis=1)
rso_decile_frame = rso_decile_frame.rename(columns={0: 'predict'})
rso_decile_hframe = h2o.H2OFrame(rso_decile_frame)
```

# Explanation at each decile

The function below displays the expected contribution of the variables relevant to the row of data with a short description

```python
def get_decile_desbribe(decile):
    row = rso_decile_frame[rso_decile_frame['row_id'] == rso_percentile_dict[decile]]
    print("This is the %dth decile of the predicted value with row_id %d, target %d, and prediction %f. Base value: %f " 
          % (decile,row['row_id'].item(),row['target'].item(),row['predict'].item(), shap_values[row.index[0],:][-1]))
    #print('\n')
    #print(row)
    if row['function'].item() == 1:
        print ("Signal generation function 1")
        return print_function1_contribs(row)
    else:
        print ("Signal generation function 2")
        return print_function2_contribs(row)
```

The function below gives the shapley plot of the choosen decile. 

```python
def get_shapley_decile(decile):
    print("Shapley plot at %dth decile" % decile)
    row = rso_decile_frame[rso_decile_frame['row_id'] == rso_percentile_dict[decile]]
    return shap.force_plot(shap_values[row.index[0],:],row[X].iloc[0])
```

The function below displays the TreeInterpreter result of choice. 

```python
def eli5_dd(decile):
    row = rso_decile_frame[rso_decile_frame['row_id'] == rso_percentile_dict[decile]]
    return eli5.explain_prediction_xgboost(rso_model,row[X].iloc[0])
```

The function below displays the LIME result of the choosen decile.

```python
explainer = LIMEExplainer(training_frame=rsovalid_pd, X=X, model=rso_model, discretize=['num1', 'num4', 'num8','num9'])

def lime(decile):
    row=rso_decile_frame[rso_decile_frame['row_id'] == rso_percentile_dict[decile]]
    explainer.explain(decile) 
```
